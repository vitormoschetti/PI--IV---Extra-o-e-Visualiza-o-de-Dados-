```python
dadosTratados = (dados['Data'].notnull())
dadosTratados = dados[dadosTratados]
dadosTratados.columns = [ 'data', 'temperaturaMedia', 'temperaturaMinima', 'temperaturaMaxima', 'precipitacao', 'finalSemana', 'consumoLitros']

fds = {
    0 : 'Dia da Semana',
    1 : 'Final de Semana'
}

dadosTratados.finalSemana = dadosTratados.finalSemana.map(fds)


# itens = dadosTratados.shape[0]

# dadosTratados.finalSemana.unique()
# meses = {
#     '*-01-': '01'
# }
# dadosTratados.data = dadosTratados.data.map(meses)
dadosTratados['temperaturaMedia'] = dadosTratados['temperaturaMedia'].str.replace(',','.')


dadosTratados['temperaturaMedia'] = dadosTratados['temperaturaMedia'].astype(float)
dadosTratados['temperaturaMedia'] = dadosTratados['temperaturaMedia'].astype(int)
dadosTratados.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,6))
sns.lineplot(y = 'consumoLitros', x = 'temperaturaMedia',
             data=dadosTratados, hue='finalSemana', 
             style='finalSemana')
plt.ylabel('Consumo de Cerveja em Litros')
plt.xlabel('Temperatura Média do Ambiente (C°)')
```

```python
sns.lmplot(y = 'consumoLitros', x = 'temperaturaMedia', data=dadosTratados, height=6)
plt.ylabel('Consumo de Cerveja em Litros')
plt.xlabel('Temperatura Média do Ambiente (C°)')
```

```python
plt.figure(figsize=[16,6])
sns.boxplot(y = 'consumoLitros', x = 'temperaturaMedia', data=dadosTratados)
plt.ylabel('Consumo de Cerveja em Litros')
plt.xlabel('Temperatura Média do Ambiente (C°)')
```

```python
plt.figure(figsize=[12,6])
sns.scatterplot(y ='temperaturaMedia', x ='consumoLitros', 
                hue='finalSemana', style='finalSemana', 
                data=dadosTratados, legend='full')
plt.ylabel('Consumo de Cerveja em Litros')
plt.xlabel('Temperatura Média do Ambiente (C°)')
```
